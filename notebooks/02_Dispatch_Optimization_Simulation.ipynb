# OneMove Intelligence: Dispatch Optimization

Simulating bipartite matching algorithms (Hungarian algorithm vs Greedy) to minimize global ETA for delivery partners.

```python
import numpy as np
from scipy.optimize import linear_sum_assignment
import time

def simulate_dispatch(num_orders, num_partners):
    # Distance matrix (ETAs in minutes)
    np.random.seed(42)
    cost_matrix = np.random.uniform(2, 20, (num_orders, num_partners))
    
    # 1. Greedy Assignment
    start = time.time()
    greedy_cost = 0
    assigned_partners = set()
    for i in range(num_orders):
        available_partners = [j for j in range(num_partners) if j not in assigned_partners]
        if not available_partners: break
        best_partner = min(available_partners, key=lambda j: cost_matrix[i, j])
        assigned_partners.add(best_partner)
        greedy_cost += cost_matrix[i, best_partner]
    greedy_time = time.time() - start
    
    # 2. Optimal Assignment (Hungarian)
    start = time.time()
    row_ind, col_ind = linear_sum_assignment(cost_matrix)
    optimal_cost = cost_matrix[row_ind, col_ind].sum()
    optimal_time = time.time() - start
    
    print(f'Greedy ETA sum: {greedy_cost:.1f} mins ({greedy_time:.4f}s)')
    print(f'Optimal ETA sum: {optimal_cost:.1f} mins ({optimal_time:.4f}s)')
    print(f'Global ETA reduction: {((greedy_cost - optimal_cost)/greedy_cost)*100:.1f}%')

simulate_dispatch(100, 150)
```
